# Philosophy Statistical project

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = pd.read_csv('philosophers2.csv')
```

```python
df = pd.DataFrame(data)
df.head()
```

## Graphique du nombre de philosophes par pays

```python
pays_count = df['Pays'].value_counts()

plt.figure(figsize=(10, 6))
pays_count.plot(kind='bar', color='skyblue')
plt.title('Nombre de Philosophes par Pays')
plt.xlabel('Pays')
plt.ylabel('Nombre de Philosophes')
plt.xticks(rotation=45)
plt.show()
```

## Graphique de la répartition des philosophes par siècle

```python
siecle_count = df['Siècle'].value_counts()

plt.figure(figsize=(10, 6))
siecle_count.plot(kind='bar', color='salmon')
plt.title('Répartition des Philosophes par Siècle')
plt.xlabel('Siècle')
plt.ylabel('Nombre de Philosophes')
plt.xticks(rotation=45)
plt.show()
```

## Graphique de la répartition des philosophes par domaine

```python
domaine_count = df['Domaine'].value_counts()

plt.figure(figsize=(40, 22))
domaine_count.plot(kind='bar', color='lightgreen')
plt.title('Répartition des Philosophes par Domaine')
plt.xlabel('Domaine')
plt.ylabel('Nombre de Philosophes')
plt.xticks(rotation=45)
plt.show()
```

## Obtenir les philosophes d'un pays

```python
def get_philosophers_by_country(country):
    # Filtrer les philosophes par pays
    philosophers_from_country = df[df['Pays'].str.lower() == country.lower()]
    
    # Vérifier si le pays a des philosophes
    if philosophers_from_country.empty:
        print(f"Aucun philosophe trouvé pour le pays {country}.")
    else:
        # Afficher le nombre de philosophes
        philosopher_count = len(philosophers_from_country)
        print(f"{country} a connu {philosopher_count} philosophe(s):")
        
        # Afficher la liste des philosophes
        print(philosophers_from_country[['Philosophe', 'Siècle', 'Domaine']])

# Demander à l'utilisateur d'entrer un pays
country_input = input("Entrez le nom d'un pays pour voir ses philosophes : ")

# Appeler la fonction pour afficher les philosophes du pays
get_philosophers_by_country(country_input)
```
